# Ejercicio 1

```python
# Creamos las variables
new_planet = ' '
planets = []

# Ciclo while
while new_planet != 'done':
    if new_planet:
        new_planet = input('Enter Planet Name')
        planets.append(new_planet)
print(planets)
```

# Ejercicio 2

```python
# Creamos las variables
new_planet = ' '
planets = []

# Ciclo while
while new_planet != 'done':
    if new_planet:
        new_planet = input('Enter Planet Name')
        planets.append(new_planet)
for planet in planets:
    print(planet)
```
